## Task 1 — Multi-Series Line Chart with Highlight

**What to build:** A line chart showing CO2 emissions over time for **all Asian countries** in the dataset, with one country highlighted.

**Requirements:**
- All countries shown (for context), but only **one highlighted in colour** — your choice which
- All other lines in grey (#DDDDDD), thinner
- Highlighted country **labelled directly** at the end of its line (not in a legend)
- Insight title that names the highlighted country and its story

> 💡 `df[df['Region'] == 'Asia']` to filter; use `go.Figure()` with a loop for per-country control


```python
import plotly.graph_objects as go

# Task 1 — Multi-series line with highlight
try:
    df
except NameError:
    import pandas as pd
    df = pd.read_csv('../data/co2_emissions.csv')

# Filter Asia
df_asia = df[df['Region'] == 'Asia']

highlight_country = "India"

fig1 = go.Figure()

# Plot all countries
for country in df_asia['Country'].unique():
    df_country = df_asia[df_asia['Country'] == country]

    if country == highlight_country:
        fig1.add_trace(go.Scatter(
            x=df_country['Year'],
            y=df_country['CO2_Mt'],
            mode='lines',
            line=dict(color="#D55E00", width=3),  # highlight color
            name=country
        ))
    else:
        fig1.add_trace(go.Scatter(
            x=df_country['Year'],
            y=df_country['CO2_Mt'],
            mode='lines',
            line=dict(color="#DDDDDD", width=1),
            showlegend=False
        ))

# Add direct label for highlighted country
df_high = df_asia[df_asia['Country'] == highlight_country]
fig1.add_annotation(
    x=df_high['Year'].max(),
    y=df_high[df_high['Year'] == df_high['Year'].max()]['CO2_Mt'].values[0],
    text=highlight_country,
    showarrow=False,
    font=dict(size=12, color="#D55E00")
)

# Layout styling
fig1.update_layout(
    title="India’s CO2 Emissions Are Rising Faster Than Most Asian Peers",
    plot_bgcolor="white",
    paper_bgcolor="white",
    font=dict(family="Arial", size=12),
    xaxis=dict(showgrid=False),
    yaxis=dict(showgrid=False),
    showlegend=False
)

fig1.show()
```

## Task 2 — Slopegraph: Regional Change 2000 vs 2022

**What to build:** A slopegraph comparing **average regional CO2 emissions** between 2000 and 2022.

**Requirements:**
- One line per region (not per country — aggregate first)
- Colour: regions that increased = one colour; decreased = another
- Values labelled at both ends of each line
- No y-axis tick labels (the endpoint labels make them redundant)
- Insight title stating which regions moved most

> 💡 `df.groupby(['Region','Year'])['CO2_Mt'].mean().reset_index()` then filter to 2000 and 2022


```python
# Task 2 — Slopegraph: regional averages
# YOUR CODE HERE
# Aggregate by Region and Year
df_region = df.groupby(['Region', 'Year'])['CO2_Mt'].mean().reset_index()

# Filter years
df_slope = df_region[df_region['Year'].isin([2000, 2022])]

fig2 = go.Figure()

for region in df_slope['Region'].unique():
    df_r = df_slope[df_slope['Region'] == region].sort_values('Year')

    y0 = df_r[df_r['Year'] == 2000]['CO2_Mt'].values[0]
    y1 = df_r[df_r['Year'] == 2022]['CO2_Mt'].values[0]

    color = "#009E73" if y1 < y0 else "#D55E00"  # green decrease, orange increase

    fig2.add_trace(go.Scatter(
        x=[2000, 2022],
        y=[y0, y1],
        mode='lines+markers',
        line=dict(color=color, width=2),
        showlegend=False
    ))

    # Left label (2000)
    fig2.add_annotation(
        x=2000,
        y=y0,
        text=f"{region} ({y0:.1f})",
        showarrow=False,
        xanchor="right",
        font=dict(size=10)
    )

    # Right label (2022)
    fig2.add_annotation(
        x=2022,
        y=y1,
        text=f"{region} ({y1:.1f})",
        showarrow=False,
        xanchor="left",
        font=dict(size=10)
    )

# Layout styling
fig2.update_layout(
    title="Most Regions Increased CO2 Emissions, With Sharpest Growth in Developing Regions",
    plot_bgcolor="white",
    paper_bgcolor="white",
    font=dict(family="Arial", size=12),
    xaxis=dict(
        tickvals=[2000, 2022],
        showgrid=False
    ),
    yaxis=dict(
        showgrid=False,
        showticklabels=False
    )
)

fig2.show()
```
